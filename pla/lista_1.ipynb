EX 0

fo z min 90*m1 + 90m2

sujeito a:
0.2*m1 + 0.3m2 >= 8000
0.2*m1 + 0.25*m2 >= 6000
0.15*m1 + 0.1*m2 >= 5000
m1,m2 >= 0

```python
import gurobipy as gp

m = gp.Model()

m1 = m.addVar(vtype='I', name="m1")
m2 = m.addVar(vtype='I', name="m2")

m.setObjective(90*m1 + 120*m2, gp.GRB.MINIMIZE)

m.addConstr(0.2*m1 + 0.3*m2 >= 8000)
m.addConstr(0.2*m1 + 0.25*m2 >= 6000)
m.addConstr(0.15*m1 + 0.1*m2 >= 5000)

m.optimize()



print(f"Valor otimo para a função objetvo: {m.objVal}")
print(f"Valores da solução: m1={m1.X}, m2={m2.X}")

```

EX00

z fo max 20*x1 + 40*x2 + 20*x3 + 15*x4 + 30*x5

sujeito a :
5*x1 + 4*x2 + 3*x3 + 7*x4 + 8*x5 <= 25
1*x1 + 7*x2 + 4*x3 + 9*x4 + 6*x5 <= 25
8*x1 + 10*x2 + 2*x3 + x4 + 10*x5 <= 25

x1,x2,x3,x4,x5 = Bool

```python
from gurobipy import Model, GRB

model = Model()

projetos = [1, 2, 3, 4, 5]
desembolsos = {
    1: [5, 1, 8],
    2: [4, 7, 10],
    3: [3, 4, 2],
    4: [7, 9, 1],
    5: [8, 6, 10]
}
retornos = {1: 20, 2: 40, 3: 20, 4: 15, 5: 30}
orcamento = [25, 25, 25]

x = model.addVars(projetos, vtype=GRB.BINARY, name="x")


#Iteração pra não precisar colocar as 3 restrições uma a uma
for ano in range(3):
    model.addConstr(sum(desembolsos[j][ano] * x[j] for j in projetos) <= orcamento[ano], name=f"Orcamento_{ano+1}")


model.setObjective(sum(retornos[j] * x[j] for j in projetos), GRB.MAXIMIZE)

model.optimize()

for j in projetos:
    if x[j].x:
        print(f"Projeto {j}")
print(f"Retorno total: {model.objVal}")

```

EX01

Caso 1 = Juros simples

fo z max = x1*0.0865 + x2*0.095 + x3*0.01 + x4*0.0875 + x5*0.0925 + x6*0.09

restricoes

x1 + x2 + x3 + x4 + x5 + x6 <= 750000
x1,x2,x3,x4,x5,x6 <= 187500
x2 + x3 + x5 <= 262500
x1,x2,x3,x4,x5,x6 >= 0

```python
import gurobipy as gp

model = gp.Model()

empresas = {1:"Acme Chemical",2:"Dyna Star",3:"Eagle Vision",4:"Micromodeling",5:"Optipro",6:"Sabre Systems"}
retornos = {1: 0.0865, 2: 0.095, 3: 0.01, 4: 0.0875, 5: 0.0925, 6: 0.09}
vencimentos = {1: 11, 2: 10, 3: 6, 4: 10, 5: 7, 6: 13}

x = model.addVars(empresas.keys(),vtype="C",name="x")

model.setObjective(sum(retornos[empresa]*x[empresa] for empresa in empresas.keys()),gp.GRB.MAXIMIZE)

model.addConstr(sum(x[j] for j in empresas.keys()) <= 750000) #Orçamento total
for empresa in empresas.keys():
    model.addConstr(x[empresa] <= 187500) #Não mais do que 25% por empresa

model.addConstr(x[1]+x[2]+x[4]+x[6] >= 375000) #Minimo 50% para as empresas 1,2,4 e 6
model.addConstr(x[2]+x[3]+x[5] <= 262500) #Não mais do que 35% para as empresas 2,3 e 5

model.optimize()

for empresa in empresas.keys():
    print(f"{empresas[empresa]}: {x[empresa].X}")

print(f"Retorno total: {sum(retornos[empresa]*x[empresa].X for empresa in empresas.keys())}")

```

Caso 2 - Juros Compostos

fo z max = x1*0.0865 + x2*0.095 + x3*0.01 + x4*0.0875 + x5*0.0925 + x6*0.09

restricoes

x1 + x2 + x3 + x4 + x5 + x6 <= 750000
x1,x2,x3,x4,x5,x6 <= 187500
x2 + x3 + x5 <= 262500
x1,x2,x3,x4,x5,x6 >= 0

```python
import gurobipy as gp

model = gp.Model()

empresas = {1:"Acme Chemical",2:"Dyna Star",3:"Eagle Vision",4:"Micromodeling",5:"Optipro",6:"Sabre Systems"}
retornos = {1: 0.0865, 2: 0.095, 3: 0.01, 4: 0.0875, 5: 0.0925, 6: 0.09}
vencimentos = {1: 11, 2: 10, 3: 6, 4: 10, 5: 7, 6: 13}

x = model.addVars(empresas.keys(),vtype="C",name="x")

model.setObjective(sum(((1+retornos[empresa])**vencimentos[empresa])*x[empresa] for empresa in empresas.keys()),gp.GRB.MAXIMIZE)

model.addConstr(sum(x[j] for j in empresas.keys()) <= 750000) #Orçamento total
for empresa in empresas.keys():
    model.addConstr(x[empresa] <= 187500) #Não mais do que 25% por empresa

model.addConstr(x[1]+x[2]+x[4]+x[6] >= 375000) #Minimo 50% para as empresas 1,2,4 e 6
model.addConstr(x[2]+x[3]+x[5] <= 262500) #Não mais do que 35% para as empresas 2,3 e 5

model.optimize()

for empresa in empresas.keys():
    print(f"{empresas[empresa]}: {x[empresa].X}")

print(f"Retorno total: {round(sum(((1+retornos[empresa])**vencimentos[empresa])*x[empresa].X for empresa in empresas.keys()),2)}")
```
