```python
import numpy as np
from scipy.optimize import newton
import matplotlib.pyplot as plt
```

```python
Re = 10000.0
karman = lambda f: 4*np.log10(Re*np.sqrt(f)) - 0.4 - (1/np.sqrt(f))
```

```python
x = np.linspace(0.002,0.008,100)
plt.plot(x,karman(x),label='Karman')    
x0 = 0.007
raiz = newton(karman,x0,full_output=True)
plt.plot(raiz[0],karman(raiz[0]),'ro',label='Raiz')
plt.legend()
plt.show()
print(raiz)
```

```python
import numpy as np
from math import e
import matplotlib.pyplot as plt
from scipy.optimize import newton
from utils import PrimeiraDerivadaCentrada,PrimeiraDerivadaProgressiva,PrimeiraDerivadaRegressiva
```

```python
f = lambda x: e**-x - x
df = lambda x: -e**-x - 1
df2 = lambda x: e**-x

x = np.linspace(-0.5,2,100)
plt.plot(x,f(x),label='f(x)')
plt.plot(x,df(x),label='df(x)')
```

```python
x0 = float(input("Digite um valor para x0: "))
secant = newton(f,x0,full_output=True)
print(secant,"\n")
newtonRaphson = newton(f,x0,fprime=df,full_output=True)
print(newtonRaphson)    

plt.plot(x,f(x),'b',label='f(x)')
plt.plot(x,[0 for _ in x],label="y=0")
plt.plot(secant[0],f(secant[0]),'ro',label='Secant')
plt.plot(newtonRaphson[0],f(newtonRaphson[0]),'go',label='Newton-Raphson')
plt.legend()
```

```python
f = lambda x : 4 + 1.8*x**2 + 1.2*x**3 - 0.3*x**4
df = lambda x : 3.6*x + 3.6*x**2 - 1.2*x**3
df2 = lambda x : 3.6 + 7.2*x - 3.6*x**2

x = np.linspace(-2,6,100)
plt.plot(x,f(x),label='f(x)')
plt.plot(x,df(x),label='df(x)')
```

```python
x0 = float(input('Digite o valor de x0: '))
secant = newton(df,x0,full_output=True)
newtonRaphson = newton(df,x0,fprime=df2,full_output=True)
plt.plot(x,[0 for _ in x],label="y=0")
plt.plot(x,df(x),'b-',label='f(x)')
plt.plot(secant[0],df(secant[0]),'ro',label='Raiz (Secante)')
plt.plot(newtonRaphson[0],df(newtonRaphson[0]),'go',label='Raiz (Newton)')
plt.legend()
plt.show()
print(secant[1],"\n")
print(newtonRaphson[1])
```
