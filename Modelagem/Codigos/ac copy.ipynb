```python
import matplotlib.pyplot as plt
from numpy import linspace      
from scipy.optimize import bisect,newton
from math import e


g = 9.81
c = 15
v = 36
t = 10

v = lambda m: ((g*m)/c) * (1 - e **(-(c/m)*t))-36


x = linspace(0,100,1000)


def regulaFalsi(f,xl,xu):
    iter = 0
    erro = 0.01
    fl = f(xl)
    fu = f(xu)

    while True:
        xr = xu-fu * (xl-xu) / (fl - fu)
        fr = f(xr)
        iter += 1
        if abs(fr) < erro:
            return xr , iter
        test = fl*fr
        if test < 0:
            xu = xr
            fu = f(xu)
        elif test > 0:
            xl = xr
            fl = f(xl)
        else:
            return xr , iter
        
raiz = regulaFalsi(v,50,65)

print(f'A raiz encontrada foi em x = {raiz[0]}\nO numero de iteracoes foi {raiz[1]}')
plt.plot(x,v(x))
plt.plot(raiz[0],v(raiz[0]),"ro")
plt.grid()
plt.show()
```
