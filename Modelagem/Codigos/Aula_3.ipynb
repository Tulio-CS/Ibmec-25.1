```python
def erroVerdadeiro(valorVerdadeiro, valorAproximado):
    return valorVerdadeiro - valorAproximado

def erroRelativoPercentualVerdadeiro(valorVerdadeiro, valorAproximado):
    return abs(round(erroVerdadeiro(valorVerdadeiro, valorAproximado) / valorVerdadeiro * 100 , 3))
```

```python
import matplotlib.pyplot as plt
from numpy import linspace, log
from utils.derivacaoNumerica import diferencialFinita
```

```python
f = lambda x: x*2 + 3*x + 5

print(diferencialFinita(f, 1))
```

```python
f = lambda x: x**3 - 0.75*x**2 +3
df = lambda x: 3*x**2 - 1.5*x
df2 = lambda x: 6*x - 1.5
df3 = lambda x: 6


x0 = 0.5
x1 = 1

x = linspace(x0, x1, 100)

plt.plot(x0, f(x0), 'ro')
plt.plot(x1, f(x1), 'ro')

taylor = f(x0) + df(x0)*(x1-x0)
taylor2 = taylor + 0.5*df2(x0)*(x1-x0)**2
taylor3 = taylor2 + (1/6)*df3(x0)*(x1-x0)**3

print(f"Erro relativo percentual verdadeiro do primeiro polinomio de taylor: {erroRelativoPercentualVerdadeiro(f(x1), taylor)}%")
print(f"Erro relativo percentual verdadeiro do segundo polinomio de taylor: {erroRelativoPercentualVerdadeiro(f(x1), taylor2)}%")
print(f"Erro relativo percentual verdadeiro do terceiro polinomio de taylor: {erroRelativoPercentualVerdadeiro(f(x1), taylor3)}%")

plt.plot(x, f(x), 'r',label='f(x)')
plt.plot(x1, taylor, 'bo',label='Taylor 1')
plt.plot(x1, taylor2, 'go',label='Taylor 2')
plt.plot(x1,taylor3, 'yo',label='Taylor 3')
plt.grid()
plt.legend()
plt.show()
```

```python
f = lambda x: log(x)
df = lambda x: 1/x
d2f = lambda x: -1/x**2
d3f = lambda x: 2/x**3

x0 = 1
x1 = 2

x = linspace(x0,x1,10)
taylor = f(x0) + df(x0)*(x1-x0)
taylor2 = taylor + 0.5*d2f(x0)*(x1-x0)**2
taylor3 = taylor2 + (1/6)*d3f(x0)*(x1-x0)**3

print(f"Erro relativo percentual verdadeiro do primeiro polinomio de taylor: {erroRelativoPercentualVerdadeiro(f(x1), taylor)}%")
print(f"Erro relativo percentual verdadeiro do segundo polinomio de taylor: {erroRelativoPercentualVerdadeiro(f(x1), taylor2)}%")
print(f"Erro relativo percentual verdadeiro do terceiro polinomio de taylor: {erroRelativoPercentualVerdadeiro(f(x1), taylor3)}%")

plt.plot(x, f(x), 'r',label='f(x)')
plt.plot(x1, f(x1), 'ro',label='f(x1)')
plt.plot(x1, taylor, 'bo',label='taylor')
plt.plot(x1, taylor2, 'go',label='taylor2')
plt.plot(x1, taylor3, 'yo',label='taylor3')
plt.legend()
plt.grid()
plt.show()
```
